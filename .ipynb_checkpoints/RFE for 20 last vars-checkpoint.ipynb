```python
trade_off_label = ["%s vars" % i for i in range(1,20)]
```

```python
from sklearn import metrics,linear_model
from sklearn.metrics import log_loss,accuracy_score, precision_score, recall_score

a = []
score = []
precision = []
recall = []
eval_df = []
for x in range(1,20):
    lr_RFE = RFE(logit,n_features_to_select=x)
    lr_RFE = lr_RFE.fit(xts,y_train)
    a.append(np.where(lr_RFE.support_ == True))
    score.append(log_loss(y_train,lr_RFE.predict_proba(xts)))
    y_pred = lr_RFE.predict(xts)
    #f1score.append(f1_score(y_train, y_pred, average='weighted'))
    precision.append(precision_score(y_train, y_pred, average='macro'))
    recall.append(recall_score(y_train, y_pred, average='macro'))
```

```python
eval_df = pd.DataFrame(np.row_stack([score,precision,recall]),
                            columns=trade_off_label,index=["log loss","precision","recall"])
display(eval_df)
xi = [i for i in range(0, len(trade_off_label))]
```

```python
import matplotlib.pyplot as plt
lscore = plt.plot(score, 'bo-', marker='o', label='Training 1-Score')
lprecision = plt.plot(precision, 'ro-', marker='o', label='Test 1-Score')
lrecall = plt.plot(recall, 'go-', marker='o', label='Test 1-Score')
plt.xticks(xi, trade_off)
plt.xlabel('Number of variables')
#plt.ylabel('Y') 
#plt.legend()
plt.show()
```
